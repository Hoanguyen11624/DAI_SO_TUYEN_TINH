```python
import numpy as np
import math

a = np.array([1, 2, 3])

# a. Dùng dot
mag_a = np.sqrt(a.dot(a))
print("Chuẩn 2 (a.dot(a)):", mag_a)

# b. Dùng toán tử @
mag_b = np.sqrt(a @ a)
print("Chuẩn 2 (a @ a):", mag_b)

# c. Dùng np.inner
mag_c = np.sqrt(np.inner(a, a))
print("Chuẩn 2 (np.inner):", mag_c)

# d. Dùng hàm lambda
mag_d = lambda x : math.sqrt(sum(i**2 for i in x))
print("Chuẩn 2 (lambda):", mag_d(a))
```

```python
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

distance = np.linalg.norm(u - v)
print("Khoảng cách giữa u và v:", distance)
```

```python
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print("Góc (radian):", theta_rad)
print("Góc (độ):", theta_deg)
```

```python
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

projection = (np.dot(u, v) / np.dot(v, v)) * v
print("Phép chiếu của u lên v:", projection)
```

```python
from PIL import Image, ImageDraw

# Mở ảnh
input_image = Image.open('C:/traidat.PNG')
input_image = input_image.convert("RGB")
input_pixels = input_image.load()

# Gauss kernel 3x3
gaussian_kernel = [[1/16, 2/16, 1/16],
                   [2/16, 4/16, 2/16],
                   [1/16, 2/16, 1/16]]

kernel = gaussian_kernel
offset = len(kernel) // 2

# Tạo ảnh mới
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

# Áp dụng làm mờ
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                for c in range(3):
                    acc[c] += pixel[c] * kernel[a][b]
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))

output_image.save('C:/traidat_lammo_gauss.PNG')
```

```python
from PIL import Image, ImageDraw

# Mở ảnh
input_image = Image.open('C:/traidat.PNG')
input_image = input_image.convert("RGB")
input_pixels = input_image.load()

# Kernel rút xương (chọn một trong hai)
# kernel = [[0, -1, 0], [-1, 5, -1], [0, -1, 0]]
kernel = [[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]

offset = len(kernel) // 2
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

# Áp dụng kernel
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                for c in range(3):
                    acc[c] += pixel[c] * kernel[a][b]
        acc = [min(max(int(i), 0), 255) for i in acc]
        draw.point((x, y), tuple(acc))

output_image.save('C:/traidat_edge.PNG')
```
